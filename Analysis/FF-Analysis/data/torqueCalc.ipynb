```python
import os
from os.path import join
AllFile = []
dest = "F:\\data\\newData"
for root, dirs, files in os.walk( dest ):
    for OneFileName in files :
        OneFullFileName = join( root, OneFileName )
        AllFile.append(OneFullFileName)
    print(AllFile)
```

```python

secondData = []
for i in AllFile:
    secondData.append(i)
```

```python
secondData
```

```python
def read_dta(dta):
    """read the dta file and return numerical array"""
    with open(dta) as f:
        position = []
        tor = []
        data  = f.readlines()
        for i in data[7:]:
              position.append(float(i[0:6]))
              tor.append(float(i[8:-1]))
        return tor,position
```

```python
import numpy as np

tor,position = read_dta(secondData[8])
```
